# Pandas with Functions

```python
hr_df=pd.read_csv('Human_Resources.csv')
hr_df
```

```python
# Let's Assume that the daily rate has increased by 10%
# Define a function that increase all clients networth by a fixed value of 10% (for simplicity sake)

def dailyrate_update(balance):
  return balance * 1.1 #assume that net worth increased by 10%
```

```python
# You can apply a function to the DataFrame

hr_df['DailyRate']=hr_df['DailyRate'].apply(dailyrate_update)
hr_df
```

# Practical of Pandas Funtion

```python
# Define a function that doubles the daily rate and adds $100
# Apply function to the DataFrame
# Calculate the updated total Daily Rate of all employees combined.
def doub_dailyrate(balance):
  return (balance * 2)+100 
```

```python
hr_df['DailyRate'] = hr_df['DailyRate'].apply(doub_dailyrate)
hr_df
```
