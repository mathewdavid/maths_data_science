```python
#Importing the required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the data

stocks_df = pd.read_csv('stocks_daily_prices.csv')
stocks_df
```

```python
#Importing the data

daily_return_df = pd.read_csv('stocks_daily_returns.csv')
daily_return_df
```

```python
# get a list of data frame names
df_names = [name for name, obj in globals().items() if isinstance(obj, pd.DataFrame)]
df_names
```

```python
#2. Plotting lineplots that will display all 4 stocks daily prices on one single figure.

stocks_df.plot(x='Date',y=['AAPL','JPM','PG','UAL'], linewidth=3,figsize=(16,8))
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend(loc='upper center')
plt.grid()
```

```python
#2. Plotting lineplots that will display all 4 stocks daily prices on one single figure.

daily_return_df.plot(x='Date', y=['AAPL','JPM','PG','UAL'], linewidth=3, figsize=(16,8))
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend(loc='upper center')
plt.grid()
```

```python
#3. Plotting 4 stocks daily prices on multiple subplots

stocks_df.plot(x='Date', title='Stock Price', subplots=True, figsize=(16,16), grid=True)
```

```python
#3. Plotting 4 stocks daily prices on multiple subplots

daily_return_df.plot(x='Date',title='Stock Price', subplots=True, figsize=(16,16), grid=True)
```

```python
#4. Plotting the 4 plots on subplots next to each other (all figures in one row).

plt.figure(figsize=(20,4))

plt.subplot(1, 4, 1)
plt.plot(stocks_df['AAPL'], 'r--')
plt.grid()

plt.subplot(1, 4, 2)
plt.plot(stocks_df['JPM'],'b.')
plt.grid()

plt.subplot(1, 4, 3)
plt.plot(stocks_df['PG'],'g.')
plt.grid()

plt.subplot(1, 4, 4)
plt.plot(stocks_df['UAL'],'y.')
plt.grid()
```

```python
#4. Plotting the 4 plots on subplots next to each other (all figures in one row).

plt.figure(figsize=(20,4))

plt.subplot(1, 4, 1)
plt.plot(daily_return_df['AAPL'], 'r--')
plt.grid()

plt.subplot(1, 4, 2)
plt.plot(daily_return_df['JPM'], 'b.')
plt.grid()

plt.subplot(1, 4, 3)
plt.plot(daily_return_df['PG'], 'g.')
plt.grid()

plt.subplot(1, 4, 4)
plt.plot(daily_return_df['UAL'], 'y.')
plt.grid()
```

```python
#5. Plotting the scatterplot with Matplotlib, between Apple and JP Morgan daily returns.

x= daily_return_df['AAPL']
y= daily_return_df['JPM']
plt.figure(figsize=(15,10))
plt.scatter(x,y, alpha=0.5);
plt.grid()
```

```python
#6. Plottting similar scatterplot with Seaborn, between Apple and JP Morgan daily returns.

plt.figure(figsize=(15,8))
sns.scatterplot(x='AAPL', y="JPM", data=daily_return_df)
plt.grid()
```

```python
from matplotlib import colors
#7. Plotting a pie chart that shows these allocations. Use 'explode’ attribute to increase the separation between AAPL and the rest of the portfolio.

values = ['10','70','10','10']
colors = ['g','r','y','b']
explode = [0, 0.2, 0, 0]
labels = ['JP Morgan Chase(JPM)', 'Apple(AAPL)', 'United Airlines (UAL)', 'P&G (PG)']

# Using matplotlib to plot a pie

plt.figure(figsize=(10,10))
plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('Stock Portfolio')
plt.show()
```

```python
# Plotting Histogram of the returns

plt.figure(figsize=(20,15))

mu_UAL = round(daily_return_df['UAL'].mean(),2) #Mean of the distribution
sigma_UAL = round(daily_return_df['UAL'].std(), 2) #Standard Deviaton of distribution

mu_PG = round(daily_return_df['PG'].mean(),2) # Mean of the following Deviation
sigma_PG = round(daily_return_df['PG'].std(),2) # Standard deviation of distribution

daily_return_df[['UAL','PG']].plot.hist(bins=60, alpha=0.7, figsize=(15,8));
plt.grid()
plt.ylabel('Probability')

plt.title('Histogram: P&G mu'+str(mu_PG)+', P&G sigma='+str(sigma_PG)+',UAL mu'+str(mu_UAL)+', UAL sigma='+str(sigma_UAL))
```

```python
# Plotting a heatmap that shows the correlations between stocks daily returns. Comment on the correlation between UAL and P&G.
plt.figure(figsize=(12,10))
cm= daily_return_df.corr()
sns.heatmap(cm, annot=True)
```

```python
# Plotting a 3D plot showing all daily returns from JPM, AAPL and UAL [External Research is required].

# Toolkits are collections of application-specific functions that extend Matplotlib.
# mpl_toolkits.mplot3d provides tools for basic 3D plotting.
# https://matplotlib.org/mpl_toolkits/index.html

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection = '3d')

x = daily_return_df['JPM'].tolist()
y = daily_return_df['AAPL'].tolist()
z = daily_return_df['UAL'].tolist()

ax.scatter(x, y, z, c = 'r', marker = 'o', s = 1000)

ax.set_xlabel('JP Morgan')
ax.set_ylabel('Apple')
ax.set_zlabel('United Airlines')
```
